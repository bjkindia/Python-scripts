```python
from metpy.cbook import get_test_data
import xarray as xr
import matplotlib.pyplot as plt
from netCDF4 import Dataset
pas = xr.open_dataset("wrf_ysu_24112020_06.nc")
#pas  =  xr.open_dataset("wrfinput_d01")
ph = pas.metpy.parse_cf('PH')
phb = pas.metpy.parse_cf('PHB')
hgt = pas.metpy.parse_cf('HGT')
tem = pas.metpy.parse_cf('T')
Pre = pas.metpy.parse_cf('P')
P_b = pas.metpy.parse_cf('PB')
Pr = (Pre+P_b)/100
tem = tem+300
z = ((ph+phb)/9.81)-hgt

# lon, lat = 47.45,21.2,
z = z.isel(Time=0).isel(south_north=282).sel(west_east=210)
tem = tem.isel(Time=0).isel(south_north=282).sel(west_east=210)
Pr = Pr.isel(Time=0).isel(south_north=282).sel(west_east=210)
T_norm = tem/(1000/Pr)**0.286



fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(T_norm,z[:-1])#, len(z))
ax.set_xlabel('Temp',fontsize=14)
ax.set_ylabel('Height (m)', fontsize=14)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# ax.fontsize=20
ax.grid()
plt.title("at lon = 47.45, lat = 21.2")
plt.ylim(0,1000)
plt.xlim(270, 280)
fig.savefig("temp_vert.png",dpi=240)
plt.show()

```
